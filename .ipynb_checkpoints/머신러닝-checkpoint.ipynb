```python
import numpy as np
import pandas as pd 

# 모델 라이브러리 선언
from sklearn import datasets, tree
from sklearn.ensemble import RandomForestRegressor
from sklearn import linear_model

# 모델 정확도 라이브러리 선언
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
featuresData = pd.read_csv('../../lee/Python_ST_EX/dataset/feature_regression_example.csv')
```

```python
featuresData.head()
```

### [실습1]
YEARWEEK, YEAR, WEEK를 int 타입으로 설정해보고 홀리데이 및 프로모션 여부 컬럼에 대해서 Y→1, N→0 컬럼을
HO_YN 컬럼 -> HOLIDAY(Y) -> 1, HOLIDAY(N) -> 0
PRO_YN -> PROMOTION(Y)->1, PROMOTION(N) -> 0 추가로 생성하세요

```python
featuresData.dtypes
```

```python
featuresData["YEARWEEK"] = featuresData["YEARWEEK"].astype(int)
featuresData["YEAR"] = featuresData["YEAR"].astype(int)
featuresData["WEEK"] = featuresData["WEEK"].astype(int)
```

```python
featuresData.head()
```

```python
featuresData["HO_YN"] = np.where(featuresData.HOLIDAY == 'Y', 1, 0)
featuresData["PRO_YN"] = np.where(featuresData.PROMOTION == 'Y', 1, 0)
```

```python
featuresData.head()
```

### [실습1] 다른풀이 방법(sklearn에서 제공하는 방법)

```python
#get_dummy, label Encoding 방법이 존재! 
```

### [실습1] 다른풀이 방법 (dicionary mapping )

```python
binarymap = {"Y":1, "N":0}
featuresData["PRO_DICT"] = featuresData.PROMOTION.map(binarymap)
```

```python
featuresData.head()
```

### [실습2] 데이터 범위를 YEARWEEK 기준 201501 ~ 201652 구간만 조회하여
FeaturesData 변수에 저장하세요

```python
featuresData = featuresData[(featuresData.YEARWEEK >= 201501) & (featuresData.YEARWEEK <= 201652)]
```

```python
print(featuresData.YEARWEEK.min())
print(featuresData.YEARWEEK.max())
```

### 지도학습-Regression

### 특성선정/ 데이터 분리

```python
corrDf = featuresData.corr()
corrDf
```

```python
# 상관관계 계수 비율 선정
# hyper parameter라는 개념으로 반복문을 돌려 각각의 지수에 따른 정확도를 구한후 가장 정확도가 높은것을 parameter로선정
featuresStd = 0.5

features = list(corrDf[abs(corrDf.QTY) > featuresStd].index)
features
```

```python
label = ["QTY"]
label
```

```python
featuresData = featuresData.sort_values("YEARWEEK")
```

```python
standardIndex = round(len(featuresData) * 0.8)
standardIndex
```

```python
featuresData.iloc[0:standardIndex,:].head()
```

```python
#년도로 오름차순 정렬했을시 80% 구간
#학습할 특성
trainingData_features =featuresData.iloc[0:standardIndex-1,][features]
#학습할 답지
trainingData_label =featuresData.iloc[0:standardIndex-1,][label]

#년도로 오름차순 정렬했을시 20% 구간
#검증할 특성
testData_features = featuresData.iloc[standardIndex:,][features]
#검증할 답지
testData_label = featuresData.iloc[standardIndex:,]
```

### https://scikit-learn.org/stable/supervised_learning.html 여러가지 다른모델 존재(링크확인)

### 모델1. DecisionTree

```python
# Define Model
model_method = tree.DecisionTreeRegressor(random_state=1)
```

```python
# Learning (Feature & Label)
model = model_method.fit(trainingData_features, trainingData_label)
```

```python
# 예측
predict = model.predict(testData_features)
predict
```

```python
predictData = pd.DataFrame(predict)
```

```python
### 컬럼명 주입
predictData.columns = ["PREDICT_DECISIONTREE"]
predictData.head()
```

```python
testData_label = testData_label.reset_index(drop = True)
testData_label.head()
```

```python
decisiontreePredict = pd.concat([testData_label,predictData], axis =1)
```

```python
decisiontreePredict.head()
```

### 모델2. Random Forest

```python
model_method2 = RandomForestRegressor(n_estimators=300, criterion='mse', n_jobs= -1, max_depth=None)
#n_estimators -> 트리개수
#max_depth -> 최대깊이
```

```python
model2 = model_method2.fit(trainingData_features, trainingData_label)
```

```python
# 예측
predict = model2.predict(testData_features)
predict
```

```python
predictData_randomforest = pd.DataFrame(predict)
```

```python
### 컬럼명 주입
predictData_randomforest.columns = ["PREDICT_RANDOMFOREST"]
predictData_randomforest.head()
```

```python
randomforestPredict = pd.concat([decisiontreePredict,predictData_randomforest],axis=1)
```

```python
randomforestPredict.head()
```

### 모델3. Multiple Regression

```python
model_method3 = linear_model.LinearRegression()
```

```python
model3 = model_method3.fit(trainingData_features, trainingData_label)
```

```python
# 예측
predict = model3.predict(testData_features)
predict
```

```python
predictData_multipleregression = pd.DataFrame(predict)
```

```python
### 컬럼명 주입
predictData_multipleregression.columns = ["PREDICT_MULTIPLEREGRESSION"]
predictData_multipleregression.head()
```

```python
finalPredict = pd.concat([randomforestPredict,predictData_multipleregression],axis=1)
```

```python
finalPredict
```

### 정확도 검증

### DecisionTree 정확도

```python
import math
```

```python
### MAE
mean_absolute_error(finalPredict['QTY'], finalPredict['PREDICT_DECISIONTREE'])
```

```python
### RMSE (MSE)
math.sqrt(mean_squared_error(finalPredict['QTY'], finalPredict['PREDICT_DECISIONTREE']))
```

```python
### R2
r2_score(finalPredict['QTY'], finalPredict['PREDICT_DECISIONTREE'])
```

### Random Forest 정확도

```python
### MAE
mean_absolute_error(finalPredict['QTY'], finalPredict['PREDICT_RANDOMFOREST'])
```

```python
### RMSE (MSE)
math.sqrt(mean_squared_error(finalPredict['QTY'], finalPredict['PREDICT_RANDOMFOREST']))
```

```python
### R2
r2_score(finalPredict['QTY'], finalPredict['PREDICT_RANDOMFOREST'])
```

### Multiple Regression 정확도

```python
### MAE
mean_absolute_error(finalPredict['QTY'], finalPredict['PREDICT_MULTIPLEREGRESSION'])
```

```python
### RMSE (MSE)
math.sqrt(mean_squared_error(finalPredict['QTY'], finalPredict['PREDICT_MULTIPLEREGRESSION']))
```

```python
### R2
r2_score(finalPredict['QTY'], finalPredict['PREDICT_MULTIPLEREGRESSION'])
```
